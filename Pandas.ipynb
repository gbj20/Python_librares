5. Find where ans_name='Scott Boston' or ans_name='Ted Petrou' or ans_name='MaxU' or ans_name='unutbu'.



```python
df4=pd.read_csv('/stackoverflow_qa.csv',index_col=0)
df4
```

```python
#1. Find all the questions that have exactly 5 answers.
answer_count=df4['answercount']==5
df4[answer_count]

```

```python
#2. Find all the questions that have less than 10 views.
view=df4['viewcount'] < 10
df4[view]
```

```python
#3. Find where quest_name='Jason Strimpel'.
res=df4['quest_name']=='Jason Strimpel'
df4[res]
```

```python
#4. Do both 1 and 2.
Final=answer_count & view
df4[Final]
```

```python
 #5. Find where ans_name='Scott Boston' or ans_name='Ted Petrou' or ans_name='MaxU' or ans_name='unutbu'.
A=df4['ans_name']=='Scott Boston'
B=df4['ans_name']=='Ted Petrou'
C=df4['ans_name']=='MaxU'
D=df4['ans_name']=='unutbu'

Z = A | B | C | D
df4[Z]
```
